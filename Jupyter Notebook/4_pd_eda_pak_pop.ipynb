## Population in Rural and Urban in Pakistan in 1998 and 2017

```python
total_rural_pop_2017 = df["ALL SEXES (RURAL)"].sum() # the rural population of Pakistan in 2017
print("The total Rural population of Pakistan in 2017 is: ",total_rural_pop_2017)
total_rural_pop_1998 = df["POPULATION 1998 (RURAL)"].sum() # the rural population of Pakistan in 1998
print("The total Rural population of Pakistan in 1998 is: ",total_rural_pop_1998)
```

```python
total_urban_pop_2017 = df["ALL SEXES (URBAN)"].sum() # the Urban population of Pakistan in 2017
print("The total Urban population of Pakistan in 2017 is: ",total_urban_pop_2017)
total_urban_pop_1998 = df["POPULATION 1998 (URBAN)"].sum() # the Urban population of Pakistan in 1998
print("The total Urban population of Pakistan in 1998 is: ",total_urban_pop_1998)
```

## Population change from 1998 to 2017

```python
# Rural population growth rate from 1998 to 2017
total_rural_pop_inc = total_rural_pop_2017 - total_rural_pop_1998
total_rural_growth_rate = (total_rural_pop_2017 - total_rural_pop_1998) * 100 / total_rural_pop_2017
print("Total rural population growth from 1998 to 2017 is: ",total_rural_pop_inc, " and the growth rate is: ",round(total_rural_growth_rate,2),"%")
```

```python
# Urban population growth rate from 1998 to 2017
total_urban_pop_inc = total_urban_pop_2017 - total_urban_pop_1998
total_urban_growth_rate = (total_urban_pop_2017 - total_urban_pop_1998) * 100 / total_rural_pop_2017
print("Total urban population growth from 1998 to 2017 is: ",total_urban_pop_inc, " and the growth rate is: ",round(total_urban_growth_rate,2),"%")
```
