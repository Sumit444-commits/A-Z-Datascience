```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
x=[1,2,3]
y=['a','b','c']
fig = px.line(x=x, y=y)
# print(fig)
fig.show()
```

```python
fig = px.bar(x=x, y=y)
# print(fig)
fig.show()
# fig.write_image("../Outputs/bar.svg")
```

```python
df= sns.load_dataset("titanic")
df.head()
```

```python
df.dropna(inplace=True)
fig2 = px.scatter(df,y="fare",x = "age",color='sex',size='fare')
fig2.show()
# save the image in 300dpi in plotly?

fig2.write_image("../Outputs/scatter_fare_age.png",scale=2,width= 800, height=600)
```
