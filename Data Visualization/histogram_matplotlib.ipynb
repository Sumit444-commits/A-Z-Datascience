# Histogram using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = np.random.randint(10,60,(50))
x
```

```python
no = [49, 21, 18, 13, 42, 50, 35, 55, 37, 24, 14, 48, 29, 31, 13, 41, 57,
       59, 41, 12, 15, 23, 22, 24, 24, 51, 16, 13, 56, 54, 33, 36, 16, 42,
       45, 17, 56, 51, 58, 18, 36, 14, 13, 10, 54, 58, 52, 35, 36, 30]
```

```python
plt.hist(no)
plt.show()
```

```python
plt.hist(no,color="b",edgecolor = 'r')
plt.show()
```

```python
bins= [10,20,30,40,50,60]
plt.hist(no,color="b",edgecolor = 'r' ,bins=bins)
plt.show()
```

```python
bins= [10,20,30,40,50,60]
plt.hist(no,color="b",edgecolor = 'r' ,bins=bins,cumulative=-1)
plt.show()
```

```python
# converting into log table
bins= [10,20,30,40,50,60]
plt.hist(no,color="b",edgecolor = 'r' ,bins=bins,log=True)
plt.show()
```

```python
# adding axvline
bins= [10,20,30,40,50,60]
plt.hist(no,color="b",edgecolor = 'r' ,bins=bins)
plt.axvline(35,color='g')
plt.show()
```
