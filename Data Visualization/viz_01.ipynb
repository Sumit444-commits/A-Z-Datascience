````python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# limport titanic dataset from seaborn library
df = sns.load_dataset("titanic")
````

## Plot

```syntax:  datasetName.plot(kind="plot type name"x, y)```

````python
df.plot(kind='scatter',x="age",y= "fare")
````

````python
df.groupby("sex")['age'].mean().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Average age")
````

## Seaborn plots

````python
# barplot
sns.barplot(data = df,x='age',y='sex')
````

````python
# barplot hue = color or grouping
sns.barplot(data = df,x='age',y='sex',hue='class')
````

Boxplot

````python
# barplot hue = color or grouping
sns.boxplot(data = df,x='age',y='sex',hue='class')
````

````python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
# z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')
z_data = df

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Titanic', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
````
