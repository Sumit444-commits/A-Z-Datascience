# heatmap

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
```

# attributes of heatmap

```python
sns.heatmap(var,vmin=0,vmax=15)
```

```python
# cmap
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr')
```

```python
# annot
sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True)
```

```python
dic = {"fontsize":10,"color":'r'}
v = sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=dic,
            linewidths=5,linecolor="g",cbar=False,xticklabels=False,
            yticklabels=False)
v.set(xlabel="python",ylabel='sumit')
plt.show()
```
