# Plotting using seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Important things for ploting according to grammer of visualiztion

1. data (df)
2. type of plot
3. x-axis
4. y-axis
5. grouping (hue/color) based on categorical variable
6. size of plot
7. color pallete
8. styles of plot
9. how to interpret

```python
df = sns.load_dataset("titanic")
```

```python
df.info()
```

```python
# scatter plot
sns.scatterplot(df,x='age',y='fare')
```

```python
sns.scatterplot(df,x='age',y='fare',hue='who')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(df,x='age',y='fare',hue='who')
```

```python
# changing color
color_palete = {
    "man":"y",
    "woman":"g",
    "child":"b"
}
plt.figure(figsize=(12,8))
sns.scatterplot(df,x='age',y='fare',hue='who',palette=color_palete)
```
