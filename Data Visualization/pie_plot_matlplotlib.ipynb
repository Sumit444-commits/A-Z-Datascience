# Pieplot using matplotlib library

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["c","c++","pyhton","Java"]
```

```python
plt.pie(x)
```

```python
plt.pie(x,labels=y)
```

```python
# explode for seperating the pie
ex = [0.0,0.0,0.1,0.0]
plt.pie(x,labels=y,explode=ex)
```

```python
# changing color
c = ["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c)
```

```python
# percentage of each pie
plt.pie(x,labels=y,explode=ex, autopct="%0.1f%%")
```

```python
plt.pie(x,labels=y,explode=ex,shadow=True)
```

```python
# increaseing the radius of pie
plt.pie(x,labels=y,explode=ex,shadow=True,radius=1.5)
```

```python
# label distance
plt.pie(x,labels=y,explode=ex,shadow=True,radius=1.5,labeldistance=0.5)
```

```python
# starting angle
plt.pie(x,labels=y,explode=ex,shadow=True,radius=1.5,labeldistance=0.5,startangle=90)
```

```python
# changing the font size of label of pie
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=0.9,radius=1.5,textprops={"fontsize":16})
```

```python
# changing the font size of percentage of pie
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=0.9,radius=1.5,textprops={"fontsize":16},pctdistance=0.6)
```

```python
# change rotation angle counter clock to clock wise
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=1.1,radius=1.5,textprops={"fontsize":16},pctdistance=0.6,counterclock=False)
```

```python
# wedge properties
wp = {"linewidth":1.2,"edgecolor":"m"}
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=1.1,radius=1.5,textprops={"fontsize":16},pctdistance=0.6,counterclock=False,wedgeprops=wp)
```

```python
# label rotate
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=1.1,
        radius=1.5,textprops={"fontsize":16},pctdistance=0.6,counterclock=False,wedgeprops=wp,
        rotatelabels=True)
```

```python
# set title
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=1.1,
        radius=1,textprops={"fontsize":16},pctdistance=0.6,counterclock=False,wedgeprops=wp,
        rotatelabels=True)
plt.title("Programming Language")
plt.show()
```

```python
# chech legend
plt.pie(x,labels=y,explode=ex,shadow=True,autopct="%0.1f%%",labeldistance=1.1,
        radius=1,textprops={"fontsize":16},pctdistance=0.6,counterclock=False,wedgeprops=wp,
        rotatelabels=True)
plt.title("Programming Language")
plt.legend()
```

## pie inside pie chart

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["c","c++","pyhton","Java"]

c = ["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x,radius=1,colors=c)
```

## Donut chart

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["c","c++","pyhton","Java"]

c = ["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
```

```python
# donut chart having 2 pie charts
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["c","c++","pyhton","Java"]

c = ["r","b","g","y"]

plt.pie(x,labels=y,radius=2)
plt.pie(x1,radius=1.5,colors=c)
plt.pie([1],colors="w")
```
