# Key Words

## Tokenization:
Tokenization is the process of breaking a stream of text up into words, phrases, symbols, or other meaningful elements. e.g. breaking a paragraph into sentences, sentences into words, etc. ("He is a boy" -> {"He","is","a","boy"})

## Stemming:
Stemming is the process of reducing inflected (or sometimes derived) words to their word stem, base or root form. e.g. "jumped","jumping"-> "jump".

## Lemmatization:

Lemmatization is the process of grouping together the inflected forms of a word so they can be analysed as a single item, identified by the word's lemma, or dictionary form. e.g. "better","best" -> "good".

## Stop Words:

Stop Words are words which are filtered out before or after processing of text. e.g. "the","is","a","an","of".
e.g: "The cat sat on a mat" -> cat, sat, mat

## Bag of words (BoW):

Bag of words is a simple way of representing text data. It involves two things: 
1. A vocabulary of known words.
2. A measure of the presence of known words.
> Tells the frequency of each token

## TF-IDF (Term Frequency-Inverse Document Frequency):

TF-IDF is a statistical measure that evaluates how relevant a word is to a document in a collection of documents. This is done by multiplying two metrics: how many times a word appears in a document, and the inverse document frequency of the word across a set of documents.
> Tells the importance of a word in a document



## NER (Named Entity Recognition):

Named Entity Recognition is a subtask of information extraction that seeks to locate and classify named entities in text into pre-defined categories such as the names of persons, organizations, locations, expressions of times, quantities, monetary values, percentages, etc.


## POS (Part of Speech):

Part of speech tagging is the process of marking up a word in a text (corpus) as corresponding to a particular part of speech, based on both its definition and its context.

## Syntax and Parsing:

Syntax is the arrangement of words and phrases to create well-formed sentences in a language. Parsing is the process of determining the syntactic structure of a text by analyzing its constituent words based on an underlying grammar.

## Smenatic Analysis:

Semantic analysis is the process of understanding the meaning of words, phrases, and sentences. It is also known as Natural Language Understanding (NLU).


## Sentiment Analysis:

Sentiment analysis is the interpretation and classification of emotions (positive, negative and neutral) within text data using text analysis techniques.

## Word Embedding:

Word embedding is the collective name for a set of language modeling and feature learning techniques in natural language processing where words or phrases from the vocabulary are mapped to vectors of real numbers.

## Sequence to Sequence Models:

Sequence to Sequence models are a class of models where the goal is to map an input sequence to an output sequence. e.g. Machine Translation, Chatbots, etc.



## Attention Mechanism:

Attention mechanism is a part of a neural network that allows the network to focus on specific parts of the input sequence when producing a certain part of the output sequence.

## Transformers:

Transformers are a type of neural network architecture that allows for significantly more parallelization than previous seq2seq models. They are based on self-attention mechanism.

e.g: BERT, GPT-2, etc.


## Language Models:

Language models are a type of model that learns to predict the probability of a word given the previous words in a sequence.

## BERT (Bidirectional Encoder Representations from Transformers):

BERT is a transformer-based machine learning technique for natural language processing pre-training developed by Google. It is designed to pre-train deep bidirectional representations from unlabeled text by jointly conditioning on both left and right context in all layers.

## GPT (Generative Pre-trained Transformer):

GPT is a series of transformer-based language models developed by OpenAI. It is designed to generate human-like text by predicting the next word in a sentence based on the previous words.

## NLP (Natural Language Processing):

Natural Language Processing is a field of artificial intelligence that focuses on the interaction between computers and humans using natural language.

## Music LLM (Music Large Language Model):

Music LLM is a type of language model that is trained on music data to generate music sequences.

# GRU

Gated Recurrent Unit (GRU) is a type of recurrent neural network that is similar to LSTM but has fewer parameters. It is designed to capture dependencies in sequential data.



# LSTM

LSTM (Long Short Term Memory) is a type of RNN that is capable of learning long-term dependencies. It is designed to overcome the vanishing gradient problem that can occur when training RNNs.



## LSTM vs GRU

LSTM and GRU are both types of RNNs that are designed to capture dependencies in sequential data. LSTM has more parameters and is capable of learning longer-term dependencies, while GRU has fewer parameters and is faster to train.
