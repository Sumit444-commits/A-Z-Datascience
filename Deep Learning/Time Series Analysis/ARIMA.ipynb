# ARIMA:
ARIMA stands for **AutoRegressive Integrated Moving Average**.
- It is a `forecasting technique` that projects the future values of a series based
entirely on its own inertia.
- Its main application is in the `area of short term forecasting` requiring at least 40
historical data points.
- It works best when your data exhibits a `stable or consistent pattern` over time
with a minimum amount of outliers
