## 2. Inter Quartile Range

```python
import matplotlib.pyplot as plt
```

```python
data = np.array([10,20,30,40,50,70,80,200])

# finding Quartiles Q1 and Q3
Q1 = np.percentile(data,25)
Q3 = np.percentile(data,75)

IQR = Q3 - Q1

# lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# finding outliers
outliers = (data < lower_bound) | (data > upper_bound)

data_with_no_outlier = data[~outliers]

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
sns.boxplot(data=data)
plt.title("With outliers")


plt.subplot(1,2,2)
sns.boxplot(data=data_with_no_outlier)
plt.title("Without outliers")

plt.show()
```

## 3. Variance

```python
data = np.array([2,3,4,5,6])

mean = np.mean(data)
# for population
pop_variance = np.var(data)

# for sample
sample_variance = np.var(data,ddof=1)

plt.figure(figsize=(10,8))
plt.bar(["Mean","Variance"],[mean,sample_variance])
plt.show()

```

## Standard Deviation

```python
data = np.array([2,3,4,5,6])

np.std(data)
```

## Standard Error

```python
std = np.std(data, ddof=1)

se = std / np.sqrt(len(data))

print("Standard Deviation : ",std)
print("Standard Error : ",se)
```
