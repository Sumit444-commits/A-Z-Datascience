# Linear Algebra Day 1
Matrix creation and solving
> Ax = B
equations:\
    2x + 3y = 8\
    5x + 4y = 20

```python
import numpy as np
```

```python
# creating matrix A
A = np.array([[2,3],[5,4]])
A
```

```python
# Creating Matrix B
B = np.array([8,20])
B
```

```python
# Now Solving for X
X = np.linalg.solve(A,B)
X
```

# Vector in python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector = np.array([3,4])
vector
```

```python
plt.figure()
plt.quiver(0,0,vector[0],vector[1],angles='xy', scale_units = "xy", scale = 1, color = 'r')
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
```

# Linear Transformation

```python
# Define the rotation matrix
theta = np.radians(45) # converting 45 degree tot radian
rotation_matrix = np.array([[np.cos(theta),-np.sin(theta)],
                            [np.sin(theta),np.cos(theta)]])
rotation_matrix
```

```python
# vector transformation
v_transform = np.dot(rotation_matrix, vector)
v_transform
```

```python
plt.figure()
plt.quiver(0,0,vector[0],vector[1],angles='xy', scale_units = "xy", scale = 1, color = 'r')
plt.quiver(0,0,v_transform[0],v_transform[1],angles='xy', scale_units = "xy", scale = 1)
plt.xlim(-1,5   )
plt.ylim(-1,5)
plt.grid()
plt.legend()
```

```python
A = np.array([[4,0],[3,-5]])
A
```
