# Data Distribution

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# load the dataset of tips
df = sns.load_dataset("tips")
df.head()
```

```python
# Making histplot 
sns.histplot(data=df, x= 'total_bill', kde=True) # kde shows line of trend
print("Right skewed distributed")
```

```python
# Making histplot 
sns.histplot(data=df, x= 'tip', kde=True) # kde shows line of trend
```

## showing distributions on titanic dataset

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Making histplot 
sns.histplot(data=titanic, x= 'survived', kde=True) # kde shows line of trend
print("bernoulli distributed")
```

```python
# Making histplot 
sns.histplot(data=titanic, x= 'age', kde=True) # kde shows line of trend
print("Right skewed distributed")
```

```python
# Making histplot 
sns.histplot(data=titanic, x= 'age', kde=True,hue='who') # kde shows line of trend
print("Poisson distributed")
```

# Kustosis and Skewness

```python
# find kurtosis
print("age kustosis",stats.kurtosis(titanic["age"] , fisher=False)) # shows nan besause of null values
print("fare kurtosis",stats.kurtosis(titanic["fare"] , fisher=False)) # 36 kurtosis means data is positively kurtosis away from normal data
```

```python
sns.histplot(data=titanic,x="fare",kde=True)
```

```python
# find skewness

print("fare kurtosis",stats.skew(titanic["fare"])) # 4.77 means data is positively skewed to right 
```

```python
# generating orignal data
np.random.seed(0)
data = np.random.normal(0,1,1000)

# calculating skewness and kurtosis
data_skew = stats.skew(data)
data_kurtosis = stats.kurtosis(data,fisher=True)

# displaying the results
print("Skewness : ",data_skew)
print("Kurtosis : ",data_kurtosis)
```

```python
sns.histplot(data=data,kde=True)
```
